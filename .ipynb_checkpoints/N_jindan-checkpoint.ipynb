```python
import pandas as pd
import numpy as np
import os, sys
import glob
```

```python
path = os.path.join("D:\code\*")

# for file in glob.glob(path):
#     print(file)
```

```python
df = pd.read_excel(
    os.path.join(".", "20220506_N대진단비 담보검토_nh.xlsx"), sheet_name='환자수',
    encoding='euc-kr', delimiter='\t')
df.head(10)
```

```python
dambo_list = list(df['코드'])
# dambo_list
```

```python
for dambo in dambo_list:
    if os.path.join("D:\code", dambo+".xls") in glob.glob(path) :
        raw_data = pd.read_excel(os.path.join("D:\code", dambo+".xls"), 
                                 encoding='euc-kr', delimiter='\t',header=6, index_col='Unnamed: 1', thousands = ',')
        
        row_condition1 = ('2020년' in raw_data.columns)
        row_condition2 = ('2021년' in raw_data.columns)
        raw_column_condition1 = (raw_data.index == '남') & (raw_data['심사년도'] == '입원')
        raw_column_condition2 = (raw_data.index == '여') & (raw_data['심사년도'] == '입원')

        if row_condition1 :
            df.loc[df['코드']== dambo, '연환자수(2020)'] = raw_data.loc['계','2020년']
        if row_condition2 :
            print(dambo)
            df.loc[df['코드']== dambo, '연환자수(2021)'] = raw_data.loc['계','2021년']
            if ('남' in raw_data.index) & ('입원' in list(raw_data.loc[raw_data.index == '남']['심사년도'])) :
#                 print(raw_data.loc[raw_column_condition1,'2021년'].values[0])
                df.loc[df['코드']== dambo, '입원_남(2021)'] = raw_data.loc[raw_column_condition1,'2021년'].values[0]

            if ('여' in raw_data.index) & ('입원' in list(raw_data.loc[raw_data.index == '여']['심사년도'])) :
#                 print(raw_data.loc[raw_column_condition2,'2021년'].values[0])
                df.loc[df['코드']== dambo, '입원_여(2021)'] = raw_data.loc[raw_column_condition2,'2021년'].values[0]
            
df.fillna('0', inplace=True)
df.replace('-','0', inplace=True)
df.replace('없음','0', inplace=True)

df['입원_남(2021)'] = df['입원_남(2021)'].astype(int)
df['입원_여(2021)'] = df['입원_여(2021)'].astype(int)

df.head(10)
```

```python
df['연환자수_입원'] = df.apply(lambda x: x['입원_남(2021)']+x['입원_여(2021)'], axis=1)
# df['연환자수_입원'] = df['입원_남(2021)'] + df['입원_여(2021)']
df
```
